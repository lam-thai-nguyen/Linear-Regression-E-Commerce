# An example of Multiple Feature Linear Regression with Gradient Descent

```python
import numpy as np

from multiple_lr import gradient_descent, _compute_cost
```

```python
# Create dataset
X = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y = np.array([460, 232, 178])
m, n = X.shape
print(f"Number of examples: {m}")
print(f"Number of features: {n}")
```

```python
# Initiate weight and bias for model
w = np.zeros(n)
b = 0
```

```python
# Initial cost
cost = _compute_cost(X, y, w, b)
print(f"Initial cost: {cost}")
```

```python
# Perform gradient descent
w_optimal, b_optimal = gradient_descent(X, y, w, b, num_iters=1000, learning_rate=5e-7, plot_cost_per_iter=True)
print(f"optimal w, b: {w_optimal}, {b_optimal}")
```
