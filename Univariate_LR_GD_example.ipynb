# An example of Univariate Linear Regression with Gradient Descent

```python
import numpy as np

from univariate_lr import gradient_descent, _compute_cost
from utils import plot_model
```

```python
# Creating a simple dataset
x = np.array([1.0, 2.0])
y = np.array([300.0, 500.0])
```

```python
# Initiate weight and bias for model
w = 0
b = 0
```

```python
# Initial cost
cost = _compute_cost(x, y, w, b)
print(f"Initial cost: {cost}")
```

```python
# Perform gradient descent
w_optimal, b_optimal = gradient_descent(x, y, w, b, num_iters=10_000, learning_rate=0.01, plot_cost_per_iter=True)
print(f"optimal w, b: {w_optimal}, {b_optimal}")
```

```python
# Plot result
plot_model(x, y, w_optimal, b_optimal)
```
